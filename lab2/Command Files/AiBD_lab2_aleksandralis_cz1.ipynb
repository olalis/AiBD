```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../Original Data/drinks.csv') #wczytanie danych oryginalnych z pliku csv

data = data.rename(columns={'country':'Country', 'beer_servings':'Beer [cans/person]', 'spirit_servings':'Spirit [shots/person]', 'wine_servings':'Wine [glasses/person]', 'total_litres_of_pure_alcohol':'Total litres of pure alcohol per person'}) #zmiana tytułów kolumn
data.to_csv('../Analysis Data/drinks_changed.csv') #zapis danych do nowego pliku

data = pd.read_csv('../Analysis Data/drinks_changed.csv') #ponowne wczytanie danych (od teraz działamy na kopii)
```

## Top 10 krajów pod względem spożycia piwa:

```python
top_of_beer = data.sort_values(by=['Beer [cans/person]'], ascending=False)[:10] #posortownie danych w tabeli 'data' po kolumnie zawterającej informacje o spożyciu piwa i wybranie 10 pierwszych rekordów
top_of_beer = top_of_beer.set_index(pd.Series([i for i in range(1, 11)])) #ustawienie nowych indeksów 
top_of_beer = top_of_beer[['Country', 'Beer [cans/person]']] #usunięcie z tabeli niepotrzbnych danych, tak aby została informacja o kraju i ilości wypitego piwa
top_of_beer.to_csv('../Analysis Data/drinks_top_of_beer.csv') #zapis tabeli do nowego pliku 
top_of_beer #wyświetlenie stworzonej tabeli z danymi
```

## Top 10 krajów pod względem spożycia wódki:

```python
top_of_spirit = data.sort_values(by=['Spirit [shots/person]'], ascending=False)[:10] #analogicznie jak w części z piwem
top_of_spirit = top_of_spirit.set_index(pd.Series([i for i in range(1, 11)]))
top_of_spirit = top_of_spirit[['Country', 'Spirit [shots/person]']]
top_of_spirit.to_csv('../Analysis Data/drinks_top_of_spirit.csv')
top_of_spirit
```

## Top 10 krajów pod względem spożycia wina:

```python
top_of_wine = data.sort_values(by=['Wine [glasses/person]'], ascending=False)[:10]  #analogicznie jak w części z piwem
top_of_wine = top_of_wine.set_index(pd.Series([i for i in range(1, 11)]))
top_of_wine = top_of_wine[['Country', 'Wine [glasses/person]']]
top_of_wine.to_csv('../Analysis Data/drinks_top_of_wine.csv')
top_of_wine
```

## Top 10 krajów pod względem ogólnego spożycia alkoholu na osobę:

```python
total = data.sort_values(by=['Total litres of pure alcohol per person'], ascending=False)[:10]  #analogicznie jak w części z piwem
total = total.set_index(pd.Series([i for i in range(1, 11)]))
total = total[['Country', 'Total litres of pure alcohol per person']]
total.to_csv('../Analysis Data/drinks_total.csv')
total
```
