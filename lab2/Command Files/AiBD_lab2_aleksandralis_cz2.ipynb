```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Top 10 krajów pod względem spożycia piwa:

```python
top_of_beer = pd.read_csv('../Analysis Data/drinks_top_of_beer.csv') #wczytanie pliku csv zawierającego przetworzone dane

fig, ax = plt.subplots()
plt.bar(top_of_beer["Country"], top_of_beer["Beer [cans/person]"]) #stworzenie wykresu słupkowego
plt.setp(ax.get_xticklabels(), rotation=45, horizontalalignment='right') #ustawienie pochylenia podpisów osi x
plt.title("Top 10 countries by serving beer") #dodanie tytułu
plt.ylim([280,380]) #ustawienie zakresu danych na osi y
plt.ylabel("cans/person") #dodanie tytułu osi y
```

### Top 10 krajów pod względem spożycia wódki:

```python
top_of_spirit = pd.read_csv('../Analysis Data/drinks_top_of_spirit.csv') #analogicznie jak w części z piwem

fig, ax = plt.subplots()
plt.bar(top_of_spirit["Country"], top_of_spirit["Spirit [shots/person]"])
plt.setp(ax.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.title("Top 10 countries by serving spirit")
plt.ylim([230,450])
plt.ylabel("shots/person")
```

### Top 10 krajów pod względem spożycia wina:

```python
top_of_wine = pd.read_csv('../Analysis Data/drinks_top_of_wine.csv') #analogicznie jak w części z piwem

fig, ax = plt.subplots()
plt.bar(top_of_wine["Country"], top_of_wine["Wine [glasses/person]"])
plt.setp(ax.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.title("Top 10 countries by serving wine")
plt.ylim([210,380])
plt.ylabel("glasses/person")
```

### Top 10 krajów pod względem ogólnego spożycia alkoholu na osobę:

```python
total = pd.read_csv('../Analysis Data/drinks_total.csv') #analogicznie jak w części z piwem

fig , ax = plt.subplots()
plt.bar(total["Country"], total["Total litres of pure alcohol per person"])
plt.setp(ax.get_xticklabels(), rotation=45, horizontalalignment='right')
plt.title("Top 10 countriesin terms of alcohol consumption")
plt.ylim([11,15])
plt.ylabel("litres/person")
```
