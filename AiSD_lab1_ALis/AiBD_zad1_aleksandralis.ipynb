```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

def fun(x):
    return x**2 + 5

x1 = np.linspace(-1, 1)
x2 = np.linspace(-6, 6)
x3 = np.linspace(0, 5)

for i in [x1, x2, x3]:
    plt.figure()
    plt.plot(i, fun(i))
    plt.title("Wykres funkcji $f(x)=x^2+5$")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.grid()
    plt.show
```
